```python
#Model Building
```

```python
X=df.drop('class',axis=1)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split 70:30

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=1)
```

```python
# KNE(kNeriest Neighbor Classifer)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=5,weights='distance')
```

```python
KNN.fit(X_train,y_train)
```

```python
predict_1=KNN.predict(X_test)
```

```python
predict_1
```

```python
from scipy.stats import zscore
print('KNN algo Accuracy :',KNN.score(X_test,y_test)*100)
```

```python
#SVM(Support Vector Machine Algo)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(X_train,y_train)
```

```python
predict_2=svc.predict(X_test)
predict_2
```

```python
print("SVM Prediction accuracy :",svc.score(X_test,y_test)*100)
```

```python
#Report
```

```python
KNN_prediction=pd.DataFrame(predict_1)
SVC_prediction=pd.DataFrame(predict_2)
```

```python
from sklearn import metrics
```

```python
print('Confusion Matrix For KNN')
print('___'*10)
cm=metrics.confusion_matrix(y_test,predict_1)
df_cm=pd.DataFrame(cm,index=[i for i in[2,4]],columns=[i for i in ['Predict B','Predict M']])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True)
```

```python
print('Confusion Matrix For SVC')
print('___'*10)
cm=metrics.confusion_matrix(y_test,predict_2)
df_cm=pd.DataFrame(cm,index=[i for i in[2,4]],columns=[i for i in ['Predict B','Predict M']])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True)
```
